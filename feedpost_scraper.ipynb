```python
from requests_html import HTMLSession
import re
import pandas as pd
# establish a session
session = HTMLSession()
```

```python
resp = session.get('https://journalists.feedspot.com/sports_news_websites/')
```

```python
links = resp.html.find("a")
```

```python
links
```

```python
import re
pattern = 'https://([^/]+).([^/]+).([^/]+)/?utm_source=feedspot'
for l in links:
  l = str(l)
  match_l = re.search(pattern,l)
  if match_l:
    res = str(match_l.group(0)) + str(match_l.group(1)) + str(match_l.group(2))
    res_l = res.split('?')
    print(res_l[0])
```

```python
def print_sites(feed_posturl):
  resp = session.get(feed_posturl)
  links = resp.html.find("a")
  pattern = 'https://([^/]+).([^/]+).([^/]+)/?utm_source=feedspot'
  for l in links:
    l = str(l)
    match_l = re.search(pattern,l)
    if match_l:
      res = str(match_l.group(0)) + str(match_l.group(1)) + str(match_l.group(2))
      res_l = res.split('?')
      print(res_l[0])
```

```python
url = 'https://journalists.feedspot.com/marketing_news_websites/'
print_sites(url)
```

```python
news_url = []
with open('/content/sample_data/news_urls.txt','r+') as file:
  for line in file:
    news_url.append(line)

urls = []
for item in news_url:
  if item not in urls:
    it_st = str(item)
    it_st = it_st.replace('\n','')
    it_st = it_st.replace('\t','')
    it_st = it_st.replace(' ','')
    urls.append(it_st)
```

```python
for item in urls:
  if item is None:
    urls.remove(item)
  elif item == '\n':
    urls.remove(item)

```

```python

with open('/content/sample_data/cleaned_news_urls.txt','w') as file:
  for ur in urls:
    file.write(ur +'\n')
```
